# Data cleaning and Manipualtion

```python
import pandas as pd
df=pd.read_csv("Ambition.csv")
```

```python
df['Head_Quarters'] = df['Head_Quarters'].str.split('+').str[0]
```

```python
df['Company_Age'] = df['Company_Age'].str.replace('years old','')
```

```python
df['reviews'] = df['reviews'][0].split(' ')[0].replace("k"," ").replace('(',' ')
```

```python
df
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df
```

```python
df.describe()
```

```python
df.reviews.unique()
```

```python
df.info()
```

```python
df['reviews'] = df['reviews'].astype('float')
```

```python
df['Company_Age'] = df['Company_Age'].astype('int')
df['rating'] = df['rating'].astype('float')
```
